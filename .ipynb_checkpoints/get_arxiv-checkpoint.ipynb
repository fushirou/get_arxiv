# 目的
1日に1回最新論文を取得し、日本語化してcsvにまとめる。  
最終的にはSlack等にあげられるか検討中
csvの項目として以下を検討中。  
題名、著者、公開された日、URL  

## 論文の取得

```python
# ! conda uninstall googletrans
# ! conda install -c conda-forge googletrans
```

```python
import arxiv
from arxiv import SortCriterion, SortOrder
import googletrans
import pandas as pd
import numpy as np
import pprint
from tqdm import tqdm
import googletrans
```

```python
search_name = "Satellite Image"
search = arxiv.Search(
    query = search_name,
#     id_list = [],
    max_results = 10,
    sort_by = SortCriterion.Relevance,
#     sort_by = SortCriterion.Relevance,
    sort_order = SortOrder.Ascending
)
```

```python
search
```

```python
# 検索結果
results = list(search.results())
```

```python
trans = googletrans.Translator()

data_list=[]
for arx_pdf in tqdm(results):
    data = []
    en2jp = trans.translate(arx_pdf.title, dest="ja", src="en").text
    data.append(arx_pdf.published.strftime("%Y-%m-%d %H:%M:%S"))
    data.append(arx_pdf.title)
    data.append(en2jp)
    data.append(arx_pdf)
    data_list.append(data)
    
#         print(arx_pdf.summary)
#         arx_pdf.download_pdf()
```

```python
column = ["公開日", "タイトル（英語）", "タイトル（日本語）", "URL"]
df = pd.DataFrame(data_list, columns=column)
df
```

```python
#　公開日で並び替え
df['公開日'] = pd.to_datetime(df['公開日'], infer_datetime_format= True)
df.sort_values(by = '公開日', ascending = False, inplace = True) 
df
```

```python
import datetime
date = datetime.datetime.now().strftime("%Y%m%d")
filename = date + "_" + search_name.replace(" ", "_") + ".csv"
df.to_csv(filename, index=None, encoding="utf-8-sig")
```
